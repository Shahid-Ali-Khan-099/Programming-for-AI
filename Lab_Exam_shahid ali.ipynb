**stopwords removal**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['news'] = df['news'].apply(remove_stopwords)
```

**Drop words if not alphabets.**

```python
def remove_non_alphabet(text):
    return ' '.join(word for word in text.split() if word.isalpha())

```

```python
df['news'] = df['news'].apply(remove_non_alphabet)
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()
def lemmitize_words(text):
    return " ".join([lemmitizer.lemmatize(word,pos='v') for word in text.split()])
```

**Vectorize using bigram and trigram techniques.**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train = [' '.join(text) if isinstance(text, list) else str(text) for text in X_train]
X_test = [' '.join(text) if isinstance(text, list) else str(text) for text in X_test]

print(X_train[:5])


cv = CountVectorizer(ngram_range=(2,2))

X_train_n_gram = cv.fit_transform(X_train)
X_test_n_gram = cv.transform(X_test)

print(f"Shape of X_train_bigram: {X_train_n_gram.shape}")
print(f"Shape of X_test_bigram: {X_test_n_gram.shape}")

rf = RandomForestClassifier(n_estimators=150)

rf.fit(X_train_n_gram,y_train)
y_pred = rf.predict(X_test_n_gram)

print ("Accuracy",accuracy_score(y_test,y_pred))
print ("Confusion matrix",confusion_matrix(y_test,y_pred))





```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train = [' '.join(text) if isinstance(text, list) else str(text) for text in X_train]
X_test = [' '.join(text) if isinstance(text, list) else str(text) for text in X_test]

print(X_train[:5])

cv = CountVectorizer(ngram_range=(3,3))

X_train_n_gram = cv.fit_transform(X_train)
X_test_n_gram = cv.transform(X_test)

print(f"Shape of X_train_trigram: {X_train_n_gram.shape}")
print(f"Shape of X_test_trigram: {X_test_n_gram.shape}")

rf = RandomForestClassifier(n_estimators=150)

rf.fit(X_train_n_gram,y_train)
y_pred = rf.predict(X_test_n_gram)

print ("Accuracy",accuracy_score(y_test,y_pred))
print ("Confusion matrix",confusion_matrix(y_test,y_pred))

```
